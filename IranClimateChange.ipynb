```python
import xarray as xr
ds = xr.open_dataset('data.nc')
ds
```

```python
Snow Coverage
```

```python
ds.isel(time=0).snowc.plot()
```

```python
ds.sel(time='2021-01-01').snowc.plot()
```

```python
SC_change = (ds.snowc.sel(time='2021-01-01')
                  - ds.snowc.sel(time='2000-01-01'))
SC_change.plot()
```

```python
mean_SC = ds.snowc.mean(dim='time')
mean_SC
```

```python
mean_SC.plot()
```

```python
(ds.snowc.mean(dim=['time', 'latitude'])).plot()
```

```python
(ds.snowc.mean(dim=['time', 'longitude'])).plot()
```

```python
ds.snowc.std(dim='time').plot()
```

```python
(ds.groupby(ds.time.dt.year)
   .mean(dim=['time', 'latitude', 'longitude'])
   .snowc
   .plot())
```

```python
(ds.groupby(ds.time.dt.month)
   .mean(dim=['time', 'latitude', 'longitude'])
   .snowc
   .plot())
```

```python
(ds.resample({'time': '10y'})
    .mean(dim=['time', 'latitude', 'longitude'])
    .snowc
    .plot()
)
```
